```r
# Read the data directly from Cloud Storage into memory.
# NOTE: Alternatively you can `gsutil -m cp {drug_74941461_path}` to copy these files
#       to the Jupyter disk.
read_bq_export_from_workspace_bucket <- function(export_path) {
  col_types <- cols(standard_concept_name = col_character(), standard_concept_code = col_character(), standard_vocabulary = col_character(), drug_type_concept_name = col_character(), stop_reason = col_character(), sig = col_character(), route_concept_name = col_character(), lot_number = col_character(), visit_occurrence_concept_name = col_character(), drug_source_value = col_character(), source_concept_name = col_character(), source_concept_code = col_character(), source_vocabulary = col_character(), route_source_value = col_character(), dose_unit_source_value = col_character())
  bind_rows(
    map(system2('gsutil', args = c('ls', export_path), stdout = TRUE, stderr = TRUE),
        function(csv) {
          message(str_glue('Loading {csv}.'))
          chunk <- read_csv(pipe(str_glue('gsutil cat {csv}')), col_types = col_types, show_col_types = FALSE)
          if (is.null(col_types)) {
            col_types <- spec(chunk)
          }
          chunk
        }))
}
dataset_74941461_drug_df <- read_bq_export_from_workspace_bucket(drug_74941461_path)

dim(dataset_74941461_drug_df)

head(dataset_74941461_drug_df, 5)
```

```r
library(lubridate)

x = dataset_74941461_drug_df %>%
    mutate(drug_exposure_start_datetime = ymd_hms(drug_exposure_start_datetime),
           drug_exposure_end_datetime = ymd_hms(drug_exposure_end_datetime)) %>%
    select(person_id, drug_exposure_start_datetime, drug_exposure_end_datetime) %>%
    group_by(person_id) %>%
    summarize(duration =
              max(drug_exposure_start_datetime, drug_exposure_end_datetime, na.rm=TRUE) -
              min(drug_exposure_start_datetime, drug_exposure_end_datetime, na.rm=TRUE))

plot(log10(as.numeric(quantile(x$duration, seq(0, 1, 0.01))) + 1))

output = x[x$duration > 3600*24*365.2425/4, "person_id", drop=FALSE]
dim(output)
```

```r
write.table(
    output,
    sep = "\t",
    na = "NA",
    row.names = FALSE,
    quote = FALSE,
    file = "lipidaemia_srWGS_withC10B_C10AX09_3months.txt")
```

```r
data = dataset_74941461_drug_df %>%
    drop_na(drug_exposure_start_datetime) %>%
    mutate(drug_exposure_start_datetime = as.Date(drug_exposure_start_datetime),
           drug_exposure_end_datetime = as.Date(drug_exposure_end_datetime)) %>%
    mutate(drug_exposure_end_datetime = pmax(drug_exposure_start_datetime+1,
                                             drug_exposure_start_datetime+days_supply,
                                             drug_exposure_end_datetime,
                                             verbatim_end_date,
                                             na.rm=TRUE)) %>%
    group_by(person_id) %>%
    arrange(drug_exposure_start_datetime, drug_exposure_end_datetime) %>%
    mutate(drug_free_preceding = as.numeric(drug_exposure_start_datetime) - cummax(lag(as.numeric(drug_exposure_end_datetime), default=-Inf)))
```

```r
sum(data$drug_free_preceding==Inf) # person_id
summary(data$drug_free_preceding)
```

```r
data = data %>%
  mutate(new_flag = is.na(drug_free_preceding) | (drug_free_preceding > 14)) %>%
  mutate(scriptseries = cumsum(new_flag))
```

```r
scriptseries = data %>%
  group_by(person_id, scriptseries) %>%
  summarize(person_id=min(person_id), start=min(drug_exposure_start_datetime), stop=max(drug_exposure_end_datetime), .groups="drop")
```

# Monitor the Cloud Analysis Environment

<div class="alert alert-block alert-success">
It is helpful to view CPU utilization and memory utilization when your Cloud analysis environment is running. For example, if you are using a Dataproc cluster and you see that the CPU utilization of the worker machines is close to 100%, you could increase the number of workers to speed up the "wall clock time" it takes to complete the analysis.
</div>

**Usage**: 

1. Copy this notebook to the workspcae you wish to check the computing resources unilization. Then run the notebook and click on the links from the output to check the utilization plots.

1. It only works for when you are the of the workspace.

1. If you are only a WRITER, and not an OWNER, of the workspace in which your Cloud analysis environment is running, these utilization plots will say "You don't have access to this data."
